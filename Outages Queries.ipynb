```python
import pandas as pd
import requests
import json
from config import api_key
```

```python
start = "2022-01-01"
end = "2022-12-31"
outageURL = f"https://api.eia.gov/v2/nuclear-outages/facility-nuclear-outages/data/?api_key={api_key}&frequency=daily&data[0]=outage&start={start}&end={end}&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=100"
outagePercentageURL = f"https://api.eia.gov/v2/nuclear-outages/facility-nuclear-outages/data/?api_key={api_key}&frequency=daily&data[0]=percentOutage&start={start}&end={end}&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000"
```

```python
outageData = requests.get(outagePercentageURL).json()
```

```python
df = pd.json_normalize(outageData['response'],'data')
df
```

```python
avg_outages = pd.DataFrame(df.groupby('facilityName')['percentOutage'].mean())
avg_outages
```

```python

avg_outages.to_csv('avg_outages_by.csv',encoding='utf-8')
```
