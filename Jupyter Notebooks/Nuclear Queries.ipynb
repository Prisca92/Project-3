```python
import pandas as pd
import requests
import json
from config import api_key
```

```python
start_year = "1960"
end_year = "2022"
prod_start_year = "1970"
prod_end_year = "2022"
#Link to number of units of nuclear generators by year.
opUnitsURL = f"https://api.eia.gov/v2/total-energy/data/?api_key={api_key}&frequency=annual&data[0]=value&facets[msn][]=NUOUPUS&start={start_year}&end={end_year}&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=100"
nuclProURL = f"https://api.eia.gov/v2/total-energy/data/?api_key={api_key}&frequency=annual&data[0]=value&facets[msn][]=NUETBUS&start={prod_start_year}&end={prod_end_year}&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=100"
genUnitsURL = f"https://api.eia.gov/v2/total-energy/data/?api_key={api_key}&frequency=annual&data[0]=value&facets[msn][]=NUGBPUS&start={prod_start_year}&end={prod_end_year}&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=100"
```
